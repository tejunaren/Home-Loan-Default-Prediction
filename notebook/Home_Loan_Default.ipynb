```python
missing_values = application.isnull().sum()

missing_values = missing_values[missing_values > 0]

print("\nMissing Values:")
print(missing_values.sort_values(ascending=False))

print("\nMissing Percentages:")
missing_percent = application.isnull().sum() / len(application) * 100
print(missing_percent[missing_percent > 0].sort_values(ascending=False))

high_missing_cols = missing_percent[missing_percent > 0.5].index

application.drop(columns = high_missing_cols,inplace = True)

print("\nDropped Columns:")
print(len(high_missing_cols))
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 5:</b> 🧹 Handle Missing Values
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, missing values are handled using appropriate preprocessing techniques. 
Numerical columns are filled using median values, while categorical columns are filled using mode values.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Columns containing excessive missing values are removed to reduce noise and improve model efficiency.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> Proper missing value treatment improves dataset consistency and enhances prediction accuracy.
</p>


```python
# -- Numerical Columns --
num_cols = application.select_dtypes(
 include=['int64', 'float64']).columns

for col in num_cols:
 application[col] = application[col].fillna(
 application[col].median())
    
# -- Categorical Columns --
cat_cols = application.select_dtypes(
 include='object').columns

for col in cat_cols:
 application[col] = application[col].fillna(
 application[col].mode()[0])
    
print("\nRemaining Missing Values:")
print(application.isnull().sum().sum())
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 10:</b> 🏦 Bureau Feature Aggregation
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, customer credit bureau data is aggregated to create summarized financial features. 
The bureau dataset contains previous credit records from external financial institutions.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Aggregation techniques such as mean, maximum, and average overdue values are used to generate meaningful customer-level financial indicators.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> Aggregated bureau features help capture customer credit behavior and improve risk prediction.
</p>

```python
bureau_agg = bureau.groupby('SK_ID_CURR').agg({
 'AMT_CREDIT_SUM':['mean', 'max'],
 'AMT_CREDIT_SUM_DEBT':['mean'],
 'CREDIT_DAY_OVERDUE':['mean']
})

bureau_agg.columns = [
 '_'.join(col).upper()
 for col in bureau_agg.columns
]

bureau_agg.reset_index(inplace=True)

# Merge Bureau Features
application = application.merge(
 bureau_agg,
 on='SK_ID_CURR',
 how='left'
)
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 11:</b> 🔗 Merge Aggregated Features
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, aggregated bureau and previous application features are merged with the main application dataset using customer identifiers.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Merging multiple datasets creates a comprehensive customer profile by combining historical financial records with current loan application information.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> Combining multiple banking datasets improves the quality and completeness of customer information.
</p>

```python
prev_agg = previous.groupby('SK_ID_CURR').agg({
 'AMT_APPLICATION':['mean'],
 'AMT_CREDIT':['mean'],
 'CNT_PAYMENT':['mean']
})

prev_agg.columns = [
 '_'.join(col).upper()
 for col in prev_agg.columns
]

prev_agg.reset_index(inplace=True)

# Merge Previous Features
application = application.merge(
 prev_agg,
 on='SK_ID_CURR',
 how='left'
)

# Fill Missing Values After Merge
application.fillna(0, inplace=True)
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 12:</b> 🎯 Feature Selection
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, input features and the target variable are separated for machine learning model training. 
The target variable represents whether a customer is likely to default on a loan.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Relevant features are selected to improve model efficiency and reduce unnecessary complexity.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> Proper feature selection helps improve model performance and reduces overfitting.
</p>

```python
X = application.drop('TARGET', axis=1)

y = application['TARGET']

print("\nFeature Shape:")
print(X.shape)

print("\nTarget Shape:")
print(y.shape)
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 13:</b> 📏 Feature Scaling
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, numerical features are standardized using StandardScaler. 
Feature scaling ensures that all variables contribute equally during machine learning model training.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Scaling is particularly important for algorithms sensitive to feature magnitude, such as Logistic Regression and Support Vector Machines.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> Feature scaling improves model stability, convergence speed, and prediction performance.
</p>

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 14:</b> ✂️ Train-Test Split
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, the dataset is divided into training and testing sets. 
The training dataset is used for model learning, while the testing dataset is used to evaluate model performance on unseen data.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Stratified splitting is applied to maintain the same class distribution in both datasets.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> Proper train-test splitting ensures fair evaluation and prevents model overfitting.
</p>

```python
X_train , X_test ,y_train ,y_test = train_test_split(
    X_scaled,
    y,
    test_size = 0.2,
    random_state = 42,
    stratify = y
)

# -- shapes of training  and testing data --
print(X_train.shape)

print(y_train.shape)

print(X_test.shape)

print(y_test.shape)
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 15:</b> 🤖 Model Building
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, multiple machine learning classification algorithms are trained using the prepared banking dataset. 
Different models are used to compare performance and identify the most suitable algorithm for loan default prediction.
</p>

<p style="font-size:16px; font-family:'Cambria';">
The following machine learning models are implemented:
</p>

<ul style="font-size:16px; font-family:'Cambria'; line-height:1.8;">
<li><b>Logistic Regression</b></li>
<li><b>Decision Tree Classifier</b></li>
<li><b>Random Forest Classifier</b></li>
<li><b>Gradient Boosting Classifier</b></li>
</ul>

<p style="font-size:16px; font-family:'Cambria';">
Hyperparameter tuning techniques are also applied to improve model accuracy and optimize performance.
</p>

```python
# -- Logistic Regression --
lr = LogisticRegression()
lr.fit(X_train ,y_train)

# -- Decision Tree with Hyperparameter Tuning --
dt_params = {
 'max_depth':[5,10],
 'min_samples_split':[2,5]
}
dt_grid = GridSearchCV(
 DecisionTreeClassifier(),
 dt_params,
 cv=3
)
dt_grid.fit(X_train, y_train)
dt = dt_grid.best_estimator_

# -- Random Forest with Hyperparameter Tuning --
rf_params = {
 'n_estimators':[100,200],
 'max_depth':[5,10]
}
rf_grid = GridSearchCV(
 RandomForestClassifier(random_state=42),
 rf_params,
 cv=3
)
rf_grid.fit(X_train, y_train)
rf = rf_grid.best_estimator_

# -- Gradient Boosting with Hyperparameter Tuning --
gb_params = {
 'n_estimators':[100,200],
 'max_depth':[5,10]
}
gb_grid = GridSearchCV(
 GradientBoostingClassifier(random_state=42),
 gb_params,
 cv=3
)
gb_grid.fit(X_train, y_train)
gb = gb_grid.best_estimator_
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 16:</b> 📊 Model Evaluation
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, all trained models are evaluated using different classification performance metrics. 
The predictions generated by each model are compared against actual target values.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Evaluation metrics such as Accuracy Score, Confusion Matrix, Precision, Recall, and F1-Score are used to measure prediction quality and classification performance.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> Model evaluation helps determine how effectively each algorithm identifies loan defaulters and non-defaulters.
</p>

```python
models = {
    'Logistic Regression': lr,
    'Decision Tree': dt,
    'Random Forest': rf,
    'Gradient Boosting': gb
}

results = {}

for name, model in models.items():
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test,y_pred)

    results[name] = acc

    print("\n",name)

    print("accuracy:",acc)

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test,y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test,y_pred))
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
 <b style="color:#1F618D;">Section 17:</b> 📉 ROC Curve and AUC Score
</h2>

<p style="font-size:16px; font-family:'Cambria';">
In this section, ROC Curves and ROC-AUC Scores are used to evaluate the classification ability of machine learning models.
</p>

<p style="font-size:16px; font-family:'Cambria';">
The ROC Curve visualizes the trade-off between True Positive Rate and False Positive Rate, while the AUC Score measures overall classification performance.
</p>

<p style="font-size:16px; font-family:'Cambria';">
Higher AUC values indicate better model capability in distinguishing loan defaulters from non-defaulters.
</p>

<p style="font-family:'Cambria'; font-size:16px;">
<b>Insight:</b> ROC-AUC analysis provides a strong measure of model classification performance and reliability.
</p>

```python
plt.figure(figsize=(8,6))

for name,model in models.items():
    y_prob = model.predict_proba(X_test)[:,1]

    fpr, tpr, _ = roc_curve(y_test, y_prob)

    auc = roc_auc_score(y_test, y_prob)

    plt.plot(
    fpr,
    tpr,
    label = f"{name} (AUC = {auc:.2f})"
    )

plt.plot([0,1],[0,1],'--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

# Save ROC Curve
plt.savefig("images/roc_curve.png", bbox_inches='tight')

plt.show()
```

<h2 style="font-family:'Book Antiqua', serif; font-size:20px;">
